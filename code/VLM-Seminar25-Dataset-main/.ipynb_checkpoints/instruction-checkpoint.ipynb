## 1. ü´Å Chest X-rays

### Dataset Structure
- **Location**: `./chest_xrays/` directory
- **Images**: `./chest_xrays/images/` contains PNG format radiographs
- **Annotations**: `./chest_xrays/annotations_len_50.json` stores disease labels and bounding boxes
- **Format**: JSON with image IDs as keys, containing:
  - `status`: "healthy" or "unhealthy"
  - `global_disease`: list of disease classifications
  - `bbox_2d`: coordinates and labels for abnormalities [x1, y1, x2, y2, label]

Below is a example to visualize one healthy and unhealthy image example.

```python
import json
import os
import random
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
import numpy as np

# ====== Configuration ======
FOLDER = './chest_xrays'
ANNOTATIONS_PATH = os.path.join(FOLDER, 'annotations_len_50.json')
IMAGES_PATH = os.path.join(FOLDER, 'images')
FONT_PATH = "/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf"  # fallback to default if not found
FONT_SIZE = 32
FIG_SIZE = (10, 5)
BASE_COLORS = ['#ff9999','#99ccff','#99ff99','#ffcc99','#cc99ff','#99ffff','#ff99ff','#ffff99']

# ====== Helper Functions ======
def draw_text(draw, pos, text, font, color, outline=2):
    for dx in [-outline, outline]:
        for dy in [-outline, outline]:
            draw.text((pos[0]+dx, pos[1]+dy), text, font=font, fill='black')
    draw.text(pos, text, font=font, fill=color)

def get_color(label, color_map):
    if label not in color_map:
        if len(color_map) < len(BASE_COLORS):
            color_map[label] = BASE_COLORS[len(color_map)]
        else:
            color_map[label] = tuple(np.random.rand(3,))
    return color_map[label]

def visualize_sample(img_id, ann, color_map):
    img_file = os.path.join(IMAGES_PATH, img_id + '.png')
    if not os.path.exists(img_file):
        print(f"Image not found: {img_file}")
        return None
    
    img = Image.open(img_file).convert("RGB")
    draw = ImageDraw.Draw(img)

    try:
        font = ImageFont.truetype(FONT_PATH, FONT_SIZE)
    except:
        font = ImageFont.load_default()

    for x1, y1, x2, y2, label in ann.get('bbox_2d', []):
        color = get_color(label, color_map)
        for w in range(6):
            draw.rectangle([x1-w, y1-w, x2+w, y2+w], outline=color)
        bbox_text = draw.textbbox((0,0), label, font=font)
        text_w = bbox_text[2] - bbox_text[0]
        text_h = bbox_text[3] - bbox_text[1]
        text_pos = (x1, y2 + 10) if y1 < 50 else (x1, y1 - text_h - 10)
        draw_text(draw, text_pos, label, font, color)

    return img

# ====== Main Execution ======
with open(ANNOTATIONS_PATH) as f:
    annotations = json.load(f)

healthy_keys = [k for k,v in annotations.items() if v['status'] == 'healthy']
unhealthy_keys = [k for k,v in annotations.items() if v['status'] == 'unhealthy']

sample_keys = []
if healthy_keys:
    sample_keys.append(random.choice(healthy_keys))
if unhealthy_keys:
    sample_keys.append(random.choice(unhealthy_keys))

color_map = {}

fig, axes = plt.subplots(1, len(sample_keys), figsize=FIG_SIZE)
if len(sample_keys) == 1:
    axes = [axes]

for i, key in enumerate(sample_keys):
    ann = annotations[key]
    img = visualize_sample(key, ann, color_map)
    if img:
        axes[i].imshow(img)
        gd = ann.get('global_disease')
        gd_text = ', '.join(gd) if gd else 'null'
        axes[i].set_title(f"ID: {key}\nStatus: {ann['status']}\nGlobal Disease: {gd_text}", fontsize=16)
        axes[i].axis('off')

plt.tight_layout()
plt.show()
```

## 2. üß† NOVA Brain MRI

### Dataset Structure
- **Location**: `./nova_brain/` directory
- **Images**: `./nova_brain/images/` contains MRI scans in standard image formats
- **Annotations**: `./nova_brain/annotations.json` contains detailed case information
- **Format**: JSON with case IDs as keys, containing:
  - `title`: Clinical case title
  - `clinical_history`: Patient background and symptoms
  - `final_diagnosis`: Confirmed medical diagnosis
  - `differential_diagnosis`: Alternative diagnostic considerations
  - `image_findings`: Dict of images with:
    - `scan_id`: Identifier for the scan
    - `caption`: Detailed radiological description
    - `bbox_2d_gold`: Gold standard bounding box coordinates [x1, y1, x2, y2]
    - `bbox_2d_raters`: Multiple expert annotations with radiologist IDs
  - `publication_date`: When the case was published
  - `link`: Reference to the original case source

Below are examples of brain MRI cases with annotations and clinical context.

```python
import json
import os
import random
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

NUM_SAMPLES = 2  # How many random cases to visualize

FOLDER = './nova_brain'
ANNOTATIONS_PATH = os.path.join(FOLDER, 'annotations.json')
IMAGES_PATH = os.path.join(FOLDER, 'images')
FONT_PATH = "/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf"
FONT_SIZE = 32

# Colors to cycle for bounding boxes (avoiding red)
COLOR_LIST = ['#ff9999','#99ccff','#99ff99','#ffcc99','#cc99ff','#99ffff','#ff99ff','#ffff99']

def draw_text(draw, pos, text, font, color='black', outline=2):
    for dx in [-outline, outline]:
        for dy in [-outline, outline]:
            draw.text((pos[0]+dx, pos[1]+dy), text, font=font, fill='white')
    draw.text(pos, text, font=font, fill=color)

def print_case_info(case_id, case_data):
    print(f"üÜî Case ID: {case_id}")
    print(f"üìù Title: {case_data.get('title', 'N/A')}")
    print(f"ü©∫ Clinical History: {case_data.get('clinical_history', 'N/A')}")
    print(f"üè• Final Diagnosis: {case_data.get('final_diagnosis', 'N/A')}")
    print(f"üîç Differential Diagnosis: {case_data.get('differential_diagnosis', 'N/A')}")
    print(f"üìÖ Publication Date: {case_data.get('publication_date', 'N/A')}")
    print(f"üîó Link: {case_data.get('link', 'N/A')}")
    print('-' * 60)

def visualize_case(case_id, case_data):
    images = case_data.get('image_findings', {})
    if not images:
        print(f"No images found for case {case_id}")
        return
    
    try:
        font = ImageFont.truetype(FONT_PATH, FONT_SIZE)
    except:
        font = ImageFont.load_default()

    n_images = len(images)
    fig, axes = plt.subplots(1, n_images, figsize=(5 * n_images, 5))
    if n_images == 1:
        axes = [axes]

    for i, (img_filename, img_info) in enumerate(images.items()):
        print(f"üñºÔ∏è {img_filename}: {img_info.get('caption', '')}")  # Print caption, no drawing
        
        img_path = os.path.join(IMAGES_PATH, img_filename)
        if not os.path.exists(img_path):
            print(f"Image not found: {img_path}")
            continue
        
        img = Image.open(img_path).convert("RGB")
        draw = ImageDraw.Draw(img)

        bboxes = img_info.get('bbox_2d_gold', [])
        for j, bbox in enumerate(bboxes):
            x1, y1, x2, y2 = bbox
            color = COLOR_LIST[j % len(COLOR_LIST)]
            draw.rectangle([x1, y1, x2, y2], outline=color, width=8)

        axes[i].imshow(img)
        axes[i].set_title(img_filename, fontsize=12)
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()

with open(ANNOTATIONS_PATH) as f:
    annotations = json.load(f)

sample_cases = random.sample(list(annotations.keys()), k=min(NUM_SAMPLES, len(annotations)))

for case_id in sample_cases:
    print_case_info(case_id, annotations[case_id])
    visualize_case(case_id, annotations[case_id]) 
```
